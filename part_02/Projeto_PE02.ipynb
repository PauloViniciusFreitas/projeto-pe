## Relação entre Area habitavel x Preço

```python
# Converter a unidade de área de pés quadrados para metros quadrados (1 pé² = 0.092903 m²)
df['GrLivArea_m2'] = df['GrLivArea'] * 0.092903

# Gráfico de dispersão com linha de regressão para analisar a relação entre Área Habitável e Preço de Venda
plt.figure(figsize=(10, 6))
sns.regplot(x=df['GrLivArea_m2'], y=df['SalePrice'], scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Relação entre Área Habitável e Preço de Venda')
plt.xlabel('Área Habitável (m²)')
plt.ylabel('Preço de Venda')
plt.show()
```
