```python
#IMPORTANDO BIBLIOTECAS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

#IMPORTANDO OS DADOS DO GITHUB
url = "https://raw.githubusercontent.com/glaucogoncalves/p-e/refs/heads/main/exames/exame-b/houses.csv"
df = pd.read_csv(url)


```

#Calculando estatísticas descritivas para variáveis numéricas, como média, mediana, desvio padrão e assimetria.



```python
from scipy.stats import skew

# Selecionar apenas colunas numéricas
numeric_df = df.select_dtypes(include=[np.number])

# Calcular estatísticas descritivas
statistics = numeric_df.describe().T
statistics["median"] = numeric_df.median()
statistics["std"] = numeric_df.std()
statistics["skewness"] = numeric_df.apply(skew)

# Exibir as estatísticas
display(statistics)
```

# Análise Estatísticas Descritivas para Variáveis Numéricas

Foram analisadas as variáveis numéricas por meio de médias, medianas, desvios-padrão e coeficientes de assimetria (skewness). A média reflete o valor central da distribuição, enquanto a mediana, menos sensível a outliers, indica o valor central mais representativo. O desvio padrão mede a dispersão dos valores em relação à média, fornecendo insights sobre a variabilidade dos preços e características estruturais das casas.

O coeficiente de assimetria (skewness) revela a direcionalidade da distribuição: valores positivos indicam distribuição assimétrica à direita, enquanto valores negativos sugerem assimetria à esquerda. No caso do conjunto de dados analisado, observa-se que variáveis como "SalePrice" e "TotalBsmtSF" apresentam altos coeficientes de skewness, sugerindo uma distribuição distorcida por valores extremos. Isso indica a necessidade de transformações, como logaritmização, para melhorar a normalidade da distribuição e permitir uma modelagem estatística mais robusta.

###Separando apenas colunas mais relevantes.


```python
# Selecionar apenas colunas relevantes
selected_columns = ["SalePrice", "GrLivArea", "TotalBsmtSF", "GarageCars"]
numeric_df = df[selected_columns]

# Calcular estatísticas descritivas
statistics = numeric_df.describe().T
statistics["median"] = numeric_df.median()
statistics["std"] = numeric_df.std()
statistics["skewness"] = numeric_df.apply(lambda x: skew(x.dropna()))

# Exibir as estatísticas
display(statistics)
```

# Análise da separção das colunas mais relevantes

Com o objetivo de reduzir a dimensionalidade e focar nas variáveis de maior impacto na precificação das casas, foram selecionadas "SalePrice" (preço de venda), "GrLivArea" (area útil acima do solo), "TotalBsmtSF" (total de metragem do porão) e "GarageCars" (quantidade de vagas de garagem). Essas variáveis apresentam relevância econômica e arquitetônica para o valor de mercado de uma propriedade.

As estatísticas descritivas dessas variáveis mostram que "GrLivArea" e "SalePrice" possuem altos valores de desvio padrão, refletindo uma grande heterogeneidade entre os imóveis. Além disso, a distribuição de "SalePrice" é fortemente assimétrica, indicando que a maioria dos imóveis tem preços mais baixos, enquanto um pequeno grupo apresenta valores extremamente elevados. Isso reforça a importância da padronização ou de técnicas estatísticas que minimizem a influência de outliers para uma análise mais precisa.

#Criando tabelas de frequência para variáveis categóricas, identificando as categorias mais frequentes e aquelas menos representadas.

```python
# Criar tabelas de frequência para variáveis categóricas
categorical_df = df.select_dtypes(include=["object"])
frequency_tables = {col: categorical_df[col].value_counts().to_frame().reset_index() for col in categorical_df.columns}

# Exibir as tabelas de frequência formatadas
print("\nTabelas de Frequência:\n")
for col, table in frequency_tables.items():
    table.columns = [col, "Frequência"]
    display(table.style.set_properties(**{'text-align': 'left'}).set_table_styles(
        [{'selector': 'th', 'props': [('text-align', 'left')]}]))
```

#Análise das Tabelas de Frequência para Variáveis Categóricas

Foram analisadas variáveis qualitativas como "MSZoning" (zona de uso do solo), "Street" (tipo de rua) e "SaleCondition" (condição da venda).

Diante disso, observou-se que a categoria "RL" (Residencial de baixa densidade) é predominante na variável "MSZoning", sugerindo que a maioria dos imóveis está localizada em zonas residenciais convencionais. Além disso, a categoria "Pave" é quase exclusiva em "Street", indicando que a maioria das residências está situada em ruas pavimentadas. A análise da "SaleCondition" mostra que a maioria das vendas ocorre sob condições normais, mas também há um subconjunto de vendas decorrentes de condições anormais, como heranças ou foreclosures.

Essas informações são úteis para a formulação de modelos preditivos, uma vez que variáveis categóricas frequentemente possuem um impacto significativo nos preços e na demanda do mercado.

#Gerar gráficos como histogramas e boxplots para investigar distribuições e possíveis outliers.

```python
# Colunas categóricas para o histograma
categorical_columns = ["MSZoning", "Street", "Utilities", "Neighborhood",
                       "BldgType", "CentralAir", "Functional", "SaleType"]

# Criar histogramas para cada categoria
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(14, 16))
axes = axes.flatten()

for i, col in enumerate(categorical_columns):
    df[col].value_counts().plot(kind='bar', ax=axes[i], color='blue', edgecolor='black')
    axes[i].set_title(f'Frequência de {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequência')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

# Análise da Visualização de Dados: Histogramas e Boxplots

Para complementar a análise exploratória, geramos histogramas e boxplots a fim de investigar a distribuição dos dados e a presença de possíveis outliers. Histogramas das variáveis categóricas, como "MSZoning", "Street" e "SaleType", revelam a frequência das categorias e permitem visualizar quais são predominantes no conjunto de dados.

Além disso, criamos boxplots para variáveis numéricas como "SalePrice", "GrLivArea" e "TotalBsmtSF", identificando outliers que podem influenciar a análise. Os boxplots evidenciaram que há um pequeno conjunto de imóveis com preços extremamente elevados, confirmando a necessidade de estratégias para lidar com esses valores atípicos.

#Relatar valores ausentes nas variáveis e sugerir estratégias para tratá-los.

```python
# Gráfico de valores faltantes
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=missing_values.index, y=missing_values.values, hue=missing_values.index, palette='Greens', legend=False)
plt.title('Valores Faltantes por Variável')
plt.xlabel('Variáveis')
plt.ylabel('Número de Valores Faltantes')
plt.xticks(rotation=90)
plt.show()
```

# Análise Valores Ausentes nas Variáveis

No conjunto de dados imobiliários, foi observado que diversas variáveis apresentam valores faltantes, sendo necessário adotar estratégias adequadas para lidar com essas ausências.

Entre as principais variáveis com valores ausentes, destacam-se aquelas relacionadas a características estruturais dos imóveis, como "LotFrontage" (frente do lote), "GarageYrBlt" (ano de construção da garagem) e "MasVnrArea" (área de alvenaria decorativa). Essas variáveis podem conter dados ausentes devido à ausência física do elemento correspondente (por exemplo, imóveis sem garagem não terão um "GarageYrBlt").

As estratégias para tratar valores ausentes incluem:

Substituição por valores estatísticos: Para variáveis numéricas, a imputação da mediana pode ser uma alternativa robusta, minimizando o impacto de outliers e preservando a distribuição dos dados.

Criação de uma categoria específica: No caso de variáveis categóricas, a substituição por uma categoria como "Desconhecido" pode ser útil para evitar a perda de informações.

Modelo de imputação: Métodos avançados, como regressão ou algoritmos de aprendizado de máquina, podem ser utilizados para prever valores ausentes com base em outras características do imóvel.

Remoção de variáveis: Caso a variável apresente uma grande quantidade de valores ausentes e sua importância seja limitada, pode-se optar por removê-la para evitar viés nos resultados.
